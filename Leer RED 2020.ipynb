```python
import csv
csvarchivo = open('CSVAristas.csv',encoding="utf8",errors='ignore')
entrada = csv.reader(csvarchivo, delimiter=",")
filas = []#filas de datos
expor = 1 #o 2
impor = 4 # o 5
cantidad = 7

for i in entrada:
    if entrada.line_num == 1:
        columnas = i
    else:
        filas.append(i)


tuplas = []
i = 1
for fila in filas:
    fila = fila[0].split(",")
    #limpiamos el string de comillas y espacios
    paisE = fila[expor].replace('"', '')
    paisE = paisE.replace(' ', '')
    #limpiamos el string de comillas
    paisI = fila[impor].replace('"', '')
    paisI = paisI.replace(' ', '')
    
    par = [paisE, paisI, float(fila[cantidad])]
    tuplas.append(par)
    
for tupla in tuplas:
    print(tupla)
    
with open('fichero.txt', 'w') as f:
    for tupla in tuplas:
        fila = tupla[0] + "," +tupla[1] + "," + str(tupla[2]) + "\n"
        f.write(fila)
        
csvarchivo.close()
```

```python
import networkx as nx  #visualize
import matplotlib.pyplot as plt 

G = nx.read_edgelist('fichero.txt',  delimiter=",", data=[('Dinero', float)], create_using=nx.DiGraph)
```

```python
plt.figure(figsize=(20,15))
nx.draw_networkx(G)
plt.show();
```

```python
import operator
# Mostramos el numero de nodos, aristas y el grado medio
N = G.number_of_nodes()
print('Nmero de nodos:', N)
L = G.number_of_edges()
print('Nmero de aristas:',L)
g_medio = 2 * L / N
print('Grado medio:',g_medio)
print("El camino minimo medio: ", nx.average_shortest_path_length(G))
#print('Grado de clustering: ', nx.average_clustering(G))
#print("Componentes conexas: ", nx.number_connected_components(G))

print("\n")
# Centralidad de grado
gCentral = nx.out_degree_centrality(G)     
print("Los 5 nodos con el indice de centralidad de grado de salida mas alto")        
ordenado1 = sorted(gCentral.items(), key=operator.itemgetter(1), reverse=True)
print(ordenado1[0])
print(ordenado1[1])
print(ordenado1[2])
print(ordenado1[3])
print(ordenado1[4])
print("-------------")

# Centralidad de grado
gCentral = nx.in_degree_centrality(G)     
print("Los 5 nodos con el indice de centralidad de grado de entrada mas alto")        
ordenado1 = sorted(gCentral.items(), key=operator.itemgetter(1), reverse=True)
print(ordenado1[0])
print(ordenado1[1])
print(ordenado1[2])
print(ordenado1[3])
print(ordenado1[4])
print("-------------")
print("\n")
```

```python
# Centralidad de intermediacin
print("Los 5 nodos con el indice de centralidad de intermediacion mas alto")
btwnCent = nx.betweenness_centrality(G)
ordenado2 = sorted(btwnCent.items(), key=operator.itemgetter(1), reverse=True)
print(ordenado2[0])
print(ordenado2[1])
print(ordenado2[2])
print(ordenado2[3])
print(ordenado2[4])
```

```python
import operator
# Centralidad de cercana
closeCent = nx.closeness_centrality(G)
ordenado3 = sorted(closeCent.items(), key=operator.itemgetter(1), reverse=True)
print("Los 5 nodos con el indice de centralidad de cercania mas alto")
print(ordenado3[0])
print(ordenado3[1])
print(ordenado3[2])
print(ordenado3[3])
print(ordenado3[4])
print("-------------")
print("\n")
```

```python
 import collections
# Pintamos la grafica de distribucion de grado de SALIDA
distrib = []
for nodo, deg in G.out_degree():
    distrib.append(deg)
deg_cnt = collections.Counter(sorted(distrib))
deg, cnt = zip(*deg_cnt.items())
cnt = tuple([x for x in cnt])
plt.figure(figsize=(8,6))
plt.bar(deg, cnt, width=0.9)
plt.plot(deg, cnt, marker='o', color='red', alpha=0.5)

plt.title("Distribucion de grado salida")
plt.ylabel("N nodos")
plt.xlabel("Grado")
plt.show()
```

```python
# Pintamos la grafica de distribucion de grado de ENTRADA
distrib = []
for nodo, deg in G.in_degree():
    distrib.append(deg)
deg_cnt = collections.Counter(sorted(distrib))
deg, cnt = zip(*deg_cnt.items())
cnt = tuple([x for x in cnt])
plt.figure(figsize=(8,6))
plt.bar(deg, cnt, width=0.9)
plt.plot(deg, cnt, marker='o', color='red', alpha=0.5)

plt.title("Distribucion de grado entrada")
plt.ylabel("N nodos")
plt.xlabel("Grado")
plt.show()
```
